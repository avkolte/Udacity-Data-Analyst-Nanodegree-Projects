## Investigation Overview

> The observation is on original loan amount, borrower's Prosper rating, loan term, borrower's stated monthly income. The factors influence Prosper rating.


## Dataset Overview

> Dataset has 81 attributes and it is consisted of borrower APRs and attributes of 113,937 loans. My main focus is on original loan amount, borrower's Prosper rating, loan term, borrower's stated monthly income these attributes.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
loan_df = pd.read_csv('prosperLoanData.csv')
```

## Distribution of Borrower APR
> The distribution of APR looks multimodal. A small peak centered at 0.1, a large peak centered at 0.2. There is also a small peak centered 0.3. Additionally, there is a very shape peak between 0.35 and 0.36. Only very few loans have APR greater than 0.43. 

```python
plt.figure(figsize=[8, 5])
#Distrubution of borrowerAPR 
sb.distplot(loan_df['BorrowerAPR'], hist=False, rug=True);
plt.xlabel('Borrower APR');
plt.title('Distribution of Borrower APR');
```

## Prosper Rating Occurences

> ProsperRating is an ordinal categorical variable with the values from higher to lower ('AA', 'A', 'B', 'C', 'D', 'E', 'HR'), ratings in data set centered at 'C' the highest 'AA' and the lowest 'HR' ratings are least.

```python
#Check last variables of interest
plt.figure(figsize=[14.70, 8.27])
plt.xticks(rotation=90)
sb.countplot(data = loan_df, x = 'ProsperRating (Alpha)', color = 'green');
plt.xticks(rotation=0);
plt.title("Distribution by Ratings".title(), fontsize = 14, weight = "bold");
plt.xlabel("Ratings");
```

## Relation between Numerical Attributes

> The relations between the numerical attributes are showed in this heat map in which we can see that having greater borrower APR causes Loan original amount to be less.  


```python
num_vars = ['LoanOriginalAmount', 'BorrowerAPR', 'StatedMonthlyIncome','BorrowerRate','ListingCategory (numeric)','Term']
cat_vars = ['Term', 'ProsperRating (Alpha)', 'EmploymentStatus']

# correlation plot
plt.figure(figsize = [8, 5])
sb.heatmap(loan_df[num_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'Purples', center = 0);

plt.title("Relation Between Numerical Attributes".title(), fontsize = 14, weight = "bold");
```

# Prosper Rating vs Loan Status

> Individuals with high rating tend to have Completed Loan Status.

```python
sb.catplot(x='Term', hue='LoanStatus', col='ProsperRating (Alpha)',
              data=loan_df, kind='count', palette='gray_r', col_wrap=4
             );

plt.suptitle("Relation Between Prosper Rating , Loan Status and Term", fontsize = 14, weight = "bold",y=1.04);
```
